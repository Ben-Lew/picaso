```python
from math import pi

import pandas as pd

from bokeh.io import show
from bokeh.models import BasicTicker, ColorBar, LinearColorMapper, PrintfTickFormatter
from bokeh.plotting import figure
from bokeh.sampledata.unemployment1948 import data
from bokeh.palettes import viridis
from bokeh.io import output_notebook
output_notebook()
```

```python
import picaso.test as test
```

```python
data1 = test.dlugach_test(output_dir='picaso.csv',rayleigh=False)
data2 = test.dlugach_test(output_dir='new_2stream.csv', approximation='4stream', rayleigh=False)
data3 = test.dlugach_test(output_dir='new_4stream.csv', approximation='4stream', stream=4, rayleigh=False)
```

```python
#read in data from picaso function
# data1 = pd.read_csv('picaso.csv')
# data2 = pd.read_csv('new.csv')
data_D = pd.read_csv('../')

#If you want to compare two different files for the heatmap just do this
# data = (data1-data2)/data2*100
```

```python
data=data1
data.columns.name = 'g0'
data['Unnamed: 0']=data['Unnamed: 0'].astype(str)
data.loc[0,'Unnamed: 0']="Ray"
data = data.set_index('Unnamed: 0')

w0s = list(data.index)#year
g0s = list(data.columns)#months

# reshape to 1D array or rates with a month and year for each row.
df = pd.DataFrame(data.stack(), columns=['albedo']).reset_index()

colors = viridis(8)
mapper = LinearColorMapper(palette=colors, low=df.albedo.min(), high=df.albedo.max())

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"

p = figure(height=300,width=300,
           x_range=w0s, y_range=list(reversed(g0s)),
           x_axis_location="above",
           tools=TOOLS, toolbar_location='below')

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "7px"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3

p.rect(x="Unnamed: 0", y="g0", width=1, height=1,
       source=df,
       fill_color={'field': 'albedo', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="12px",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')
p.axis.major_label_text_font_size='12px'
show(p)      # show the plot
```
